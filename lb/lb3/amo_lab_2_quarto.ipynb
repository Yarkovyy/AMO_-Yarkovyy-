Асимптотичної складність алгоритму сортування методом вставляння дорівнює $O(n^2)$.  
Складність сортування бульбашкою в найгіршому випадку також $O(n^2)$.

## Виконання коду

Сортування вставлянням:

```python
import random
from array import array
import time

SZ = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]

I1 = array('i')
for i in range(5):
    I1.append(random.randint(10,20))
    
I2 = array('i')
for i in range(10):
    I2.append(random.randint(10,20))
    
I3 = array('i')
for i in range(50):
    I3.append(random.randint(10,20))
    
I4 = array('i')
for i in range(100):
    I4.append(random.randint(10,20))
    
I5 = array('i')
for i in range(500):
    I5.append(random.randint(10,20))

I6 = array('i')
for i in range(1000):
    I6.append(random.randint(10,20))


I7 = array('i')
for i in range(2000):
    I7.append(random.randint(10,20))


I8 = array('i')
for i in range(3000):
    I8.append(random.randint(10,20))

I9 = array('i')
for i in range(4000):
    I9.append(random.randint(10,20))


I10 = array('i')
for i in range(5000):
    I10.append(random.randint(10,20))



    


Tinsert=array('d')
# 

# Сорутвання методом вставок (включення)


def insertion_sort(nums):
    # починаємо з другого елемент
    for j in range(1, len(nums)):

        key = nums[j]
        # Зберігаємо посилання на індекс першого елемента
        i = j - 1
        # Елементи відсортованого сегрмента переміщуємо вперед, якщо вони більше елемента вставки
        while i >= 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i -= 1
        # Вставляємо елемент
        nums[i + 1] = key


start_time = time.time()
insertion_sort(I1)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I2)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I3)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I4)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I5)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I6)
t=(time.time() - start_time)
Tinsert.append(t)


start_time = time.time()
insertion_sort(I7)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I8)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I9)
t=(time.time() - start_time)
Tinsert.append(t)

start_time = time.time()
insertion_sort(I10)
t=(time.time() - start_time)
Tinsert.append(t)



print("--- %s seconds ---" % (Tinsert))
```

Сортування бульбашкою:

```python
import random
from array import array
import time

def bubble_sort(array):
    length = len(array)
    for i in range(0, length):
        for j in range(0, length - i - 1):
            if array[j] > array[j + 1]:
                temp = array[j]
                array[j] = array[j + 1]
                array[j + 1] = temp



B1=array('i')
for i in range(5):
    B1.append(random.randint(10,20))
    
B2=array('i')
for i in range(10):
    B2.append(random.randint(10,20))
    
B3=array('i')
for i in range(50):
    B3.append(random.randint(10,20))

B4=array('i')
for i in range(100):
    B4.append(random.randint(10,20))


B5=array('i')
for i in range(500):
    B5.append(random.randint(10,20))


B6=array('i')
for i in range(1000):
    B6.append(random.randint(10,20))


B7=array('i')
for i in range(2000):
    B7.append(random.randint(10,20))


B8=array('i')
for i in range(3000):
    B8.append(random.randint(10,20))


B9=array('i')
for i in range(4000):
    B9.append(random.randint(10,20))


B10=array('i')
for i in range(5000):
    B10.append(random.randint(10,20))


Tbuble=array('d')

start_time = time.time()
bubble_sort(B1)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B2)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B3)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B4)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B5)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B6)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B7)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B8)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B9)
t=(time.time() - start_time)
Tbuble.append(t)

start_time = time.time()
bubble_sort(B10)
t=(time.time() - start_time)
Tbuble.append(t)


print("--- %s seconds ---" % (Tbuble))
```
